```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
# 1. Load the data
df = pd.read_csv('advertisement_clicks.csv')
```

```python
df.head()
```

```python
# 2. Separate the groups
a = df[df['advertisement_id'] == 'A']['action']
b = df[df['advertisement_id'] == 'B']['action']
```

```python
# 3. Calculate the CTR (Mean)
n_a, n_b = len(a), len(b)
clicks_a, clicks_b = a.sum(), b.sum()
ctr_a, ctr_b = a.mean(), b.mean()

print(f"Group A: {clicks_a} clicks out of {n_a} (CTR: {ctr_a:.4f})")
print(f"Group B: {clicks_b} clicks out of {n_b} (CTR: {ctr_b:.4f})")
```

```python
# 4. Manual Z-test (Formula from the lecture)
# Calculate pooled proportion
p_combined = (clicks_a + clicks_b) / (n_a + n_b)

# Calculate Standard Error (SE)
se = np.sqrt(p_combined * (1 - p_combined) * (1/n_a + 1/n_b))

# Calculate Z-score
z_score = (ctr_a - ctr_b) / se

# Calculate P-value (Two-sided test)
p_value = stats.norm.sf(abs(z_score)) * 2

print("\n--- Results ---")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.10f}")

if p_value < 0.05:
    print("Result: STATISTICALLY SIGNIFICANT. One ad is performing better.")
else:
    print("Result: NOT SIGNIFICANT. The difference is likely due to noise.")
```

```python
# 4. Alternative: Perform the test using Scipy functions
# Calculate the Z-score and P-value
z_score, p_value = stats.ranksums(a, b)

# OR using a T-test (very similar to Z-test for large N):
# t_stat, p_value = stats.ttest_ind(a, b)

print("\n--- Scipy Function Results ---")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.10f}")

if p_value < 0.05:
    print("Result: STATISTICALLY SIGNIFICANT. One ad is performing better.")
else:
    print("Result: NOT SIGNIFICANT. The difference is likely due to noise.")
```

```python
# 4. Alternative: Perform the test using Scipy functions
# Use a T-test (The T-test results are nearly identical to a Z-test for large N)
t_stat, p_value = stats.ttest_ind(a, b)

print("\n--- Scipy Function Results ---")
# Use t_stat here to match the variable above
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.10f}")

if p_value < 0.05:
    print("Result: STATISTICALLY SIGNIFICANT. One ad is performing better.")
else:
    print("Result: NOT SIGNIFICANT. The difference is likely due to noise.")
```
